```python
no_nan_n_df = nursing_df.dropna(subset = ['Location'])
no_nan_n_df.count()
```

```python
coord_n_df = no_nan_n_df.copy()
coord_n_df['Coordinates'] = [re.search('\([-]?\d+\.\d+, [-]?\d+\.\d+\)', row).group() if len(re.findall('\([-]?\d+\.\d+, [-]?\d+\.\d+\)', row)) > 0 else 'Null' for row in no_nan_n_df['Location'] ]
# q = re.findall('\(', nursing_df['Location'][1])
# print(nursing_df['Location'][1])
# print(q.head())
print(len(coord_n_df))
coord_n_df.head(10)
```

```python
coord_n_df = coord_n_df.loc[coord_n_df['Coordinates'] != 'Null']
print(len(coord_n_df))
coord_n_df.head(10)
```

```python
coord_n_df['Latitude'] = [re.search('[-]?\d+\.\d+', row).group() for row in coord_n_df['Coordinates']]
coord_n_df.head()
```

```python
coord_n_df['Longitude'] = [re.findall('[-]?\d+\.\d+', row)[1] for row in coord_n_df['Coordinates']]
print(coord_n_df.shape)
coord_n_df.head()
```

```python
del coord_n_df['Coordinates']
len(coord_n_df.columns)
```

```python
columns_to_keep = [
    "index",
    "Provider Name",
    "Provider State",
    "Federal Provider Number",
    "Provider SSA County Code",
    "Provider County Name",
    "Overall Rating",
    "Location",
    "Latitude",
    "Longitude"
]

```

```python
# create new data for coordinates with FPN as primary key
coord_df = coord_n_df[columns_to_keep].copy()

coord_df
```

```python
rating_df = coord_df[coord_df['Overall Rating'] == 5 ]

rating_df
```

```python
# save and export csv file to Resources folder 
homes_5rating = pd.DataFrame(rating_df)

homes_5rating.to_csv('homes_5stars.csv', index=False)
```
