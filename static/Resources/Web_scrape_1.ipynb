```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit the site
url = 'https://medicareguide.com/best-states-for-elderly-healthcare-340840'
browser.visit(url)
# Optional delay for loading the page
browser.is_element_present_by_css('div.tablesorter-header-inner', wait_time=1)
```

```python
# Scrape the website
html = browser.html

# Create a BeautifulSoup object from the scraped HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Find the table
table = soup.find('table', class_='tablesorter tablesorter-default')
```

```python
# Create an empty list
columns = []


#Get the columns names
column_names = table.find_all('strong')

for column_name in column_names:
    title = column_name.text
    columns.append(title)

print(columns)
```

```python
# Loop through the scraped data to create a list of rows
data = []

rows = soup.find_all('tr', {'role': 'row'})

for row in rows:
    items = []
    row_data = row.find_all('td')
    for result in row_data:
        items.append(result.text)
    data.append(items)
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
healthcare_df = pd.DataFrame(data, columns = columns)
healthcare_df
```

```python
# Delete the first row (index 0)
healthcare_df = healthcare_df.drop(0, axis=0)
healthcare_df.head()
```

```python
# Export to csv file
healthcare_df.to_csv('healthcare_rank_for_elderly.csv', index=False, sep=',', encoding='utf-8', header=True)
```

```python
browser.quit()
```
