```python
# Import python modules and packages
import csv
import pandas as pd
```

```python
# Read in the raw data from the .csv file for the Healthcare and Taxes
healthcare_df = pd.read_csv('healthcare_rank_for_elderly.csv')
taxes_df = pd.read_csv('Property-Taxes_by_State.csv')

# Verify the data was imported
#print(f'Shape of DataFrame: {nursing_df.shape}')
healthcare_df.head()
```

```python
#Investigate data types
healthcare_df.dtypes
```

```python
#Rename columns
healthcare_df.rename(columns = {'Overall Rank':'Healthcare Rank',
                               "Score": "Healthcare Score",
                               "Cost": "Healthcare Cost",
                               "Quality": "Healthcare Quality",
                               "Access": "Healthcare Access"}, inplace = True)
healthcare_df.head()
```

```python
#Investigate data types
taxes_df.head()
```

```python
taxes_df.dtypes
```

```python
#Rename Rank column
taxes_df.rename(columns = {'Rank':'Tax Rank'}, inplace = True)
taxes_df.head()
```

```python
#Merge the two data frames
combined_df = pd.merge(healthcare_df, taxes_df, on="State")
combined_df.head()
```

```python
#Rename State column
combined_df.rename(columns = {'State':'Provider State'}, inplace = True)
combined_df.head()
```

```python
#Sort by State in Ascending Order
combined_df = combined_df.sort_values(by=['Provider State'])
combined_df.head(10)
```

```python
# Save data to csv and json files
combined_df.to_csv('Healthcare_Tax_Merge_Table.csv', index=False)
combined_df.to_json('Healthcare_Tax_Merge_Table.json')
```
